```python
import pandas as pd
import numpy as np
```

```python
#primero creamos una lista que contenga en si toda las rutas de acceso a los archivos csv
lista=[ f'datos_IMECA/csv/imeca{2005+i}.csv' for i in range(0,16)]
lista
```

```python
#Realizamos la lectura de todos los archivos csv juntos
DF_list = [(pd.read_csv(f'{i}',sep=',')).replace(-99, np.nan) for i in lista]
```

```python
#comprobamos que se puedan leer de forma correcta los datos
DF_list[0]
```

```python
DF_list[13]
```

```python
#el Dataframe número 14 y 15 tienen 32 columnas cuando lo correcto serían 27
#Eliminamos las columnas que sobran
cols = [7,13,19,25,31]
DF_list[14]=DF_list[14].drop(DF_list[14].columns[cols],axis=1)
DF_list[15]=DF_list[15].drop(DF_list[15].columns[cols],axis=1)
```

```python
indices=( "Fecha"           ,        "Hora"                ,          "Noroeste.Ozono"               ,
 "Noroeste.dioxido.de.azufre"   , "Noroeste.dioxido.de.nitrogeno", "Noroeste.monoxido.de.carbono" ,
 "Noroeste.PM10"                , "Noreste.Ozono"                , "Noreste.dioxido.de.azufre"    ,
 "Noreste.dioxido.de.nitrogeno" , "Noreste.monoxido.de.carbono"  , "Noreste.PM10"                 ,
 "Centro.Ozono"                 , "Centro.dioxido.de.azufre"     , "Centro.dioxido.de.nitrogeno"  ,
 "Centro.monoxido.de.carbono"   , "Centro.PM10"                  , "Suroeste.Ozono"               ,
 "Suroeste.dioxido.de.azufre"   , "Suroeste.dioxido.de.nitrogeno", "Suroeste.monoxido.de.carbono" ,
 "Suroeste.PM10"                , "Sureste.Ozono"                , "Sureste.dioxido.de.azufre"    ,
 "Sureste.dioxido.de.nitrogeno" , "Sureste.monoxido.de.carbono"  , "Sureste.PM10" )
```

```python
#como se puede observar existe un pequeños inconveniente, por alguna razón el DataFrame crea columnas de más
# las eliminaremos,
#Segundo inconveniente: según la documentación entregada, los valores -99, representan los datos de tipo nan en esta dataset
DF_list[0].isna().sum()
```

```python
#ahora eliminamos las columnas que aparecen como sobrantes
DF_list2=[i.dropna(axis=1, how='all') for i in DF_list]
DF_list2[0]
```

```python
 #Revisamos que todos los DataFrame tengan el mismo tamaño
for i in range(0,16):
    print(np.shape(DF_list2[i]))
```

```python
#Renombramos todas las columnas para no generar problemas por los nombres de las mismas
#DF_list2[0].columns
#indices
for i in DF_list2:
    i.columns = indices

```

```python
#observamos que las columnas sobrantes se fueron
DF_list2[15].isna().sum()
```

```python

DF_all = pd.concat([i for i in DF_list2],axis=0)
DF_all
```

```python
DF_all.isna().sum()
```

```python
#Separación de lainformación por tipo de contenido
#O3 <-    DF_save[0]
#SO2 <- - DF_save[1]
#NO2 <- - DF_save[2]
#CO <- -  DF_save[3]
#PM10 <- -DF_save[4]
DF_save = list()
palabra = ['Ozono','dioxido.de.azufre','dioxido.de.nitrogeno','monoxido.de.carbono','PM10']
for i in range(0,5):
    DF_save.append(DF_all[['Fecha','Hora',f'Noreste.{palabra[i]}',f'Suroeste.{palabra[i]}',f'Noroeste.{palabra[i]}',f'Sureste.{palabra[i]}',f'Centro.{palabra[i]}']])
```
